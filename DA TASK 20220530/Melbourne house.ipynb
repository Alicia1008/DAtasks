```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
#Since a property will at least can park their car offstreet, we will fill in N/A in Car column with 1
```

```python
df['Car']=df['Car'].fillna('1')
```

```python
#For easy number reading, property price will be shown in $Million
```

```python
df['Price']=(df['Price']/1000000).round(2)
df.rename(columns = {'Price':'Price in $M'}, inplace = True)
```

```python
#since postcode,number of bedroom/bathroom, number of carpark and landsize are all integer number, the data type of these column will change to integer
```

```python
df[['Postcode','Bedroom2','Bathroom','Car','Landsize']]=df[['Postcode','Bedroom2','Bathroom','Car','Landsize']].astype(int)
```

```python
#South-Eastern Metropolitan and Southern Metropolitan area has the highest price sold.From the two graphs above, we will be focusing on Southern Metropolitan region which has the highest selling price on house and most numbers sold.
```

```python
##4. now let's look in details which suburb in this Southern Metropolitan region has the highest price and number of house sold
```

```python
#Bentleigh East, Balwyn North and Glen Iris sold most properties in Southern Metro region.Now we shall look at which suburb has the highest selling price in these suburbs.
```

```python
sns.set(rc={"figure.figsize":(10, 5)}) 
sns.barplot(data=df_SouthP.nlargest(5, 'Price in $M'),x = 'Suburb',y = 'Price in $M',color = 'blue')
plt.xticks(rotation = 65)
```

```python
#Canterbury, Hawthorn and Kew has the highest selling house price in souther metro retion.
```

```python
#Let's take a look in above three suburb, if there is any correlation between selling price and other factorss.
```

```python
df_CHK=df_H.loc[df_H['Suburb'].isin(['Canterbury','Hawthorn','Kew'])]
df_CHK
```

```python
sns.set(rc={"figure.figsize":(40, 8)})
sns.stripplot(x='Distance', y='Price in $M', data=df_H)
plt.title("R/ship b/w house price and distance to City", fontsize = 20) # for title
plt.xlabel("Distance to City", fontsize = 15) # label for x-axis
plt.ylabel("House Price", fontsize = 15) # label for y-axis
plt.xlim(1,1000)
plt.show()
```

```python
# we can see that from above graph, there's close r/ship between house price and distance to City. The closer the suburb is to City, the higher selling price might be. This will explain part of the reasons why Canterbury, Hawthorn and Kew can sell the highest prices for house.
```

```python
# Now let's see if there's any r/ship between the landsize and selling price
```

```python
sns.set(rc={"figure.figsize":(35, 8)})
sns.stripplot(x='Landsize', y='Price in $M', data=df_H)
plt.title("R/ship b/w house price and Landsize", fontsize = 20) # for title
plt.xlabel("Landsize", fontsize = 15) # label for x-axis
plt.ylabel("House Price", fontsize = 15) # label for y-axis
plt.xlim(1,1000)
plt.show()
```

```python
sns.set(rc={"figure.figsize":(65, 8)})
sns.lmplot(x = "Landsize", y = "Price in $M", data = df_CHK,height=3.5, aspect=5.5)
plt.xlim(1,1000) #as from the beginning of this study, we know that 75% of the land size are under 675, here we take 1000 as the upper limit for data
plt.show()
```

```python
#We can conclude that from above study that the bigger the landsize, the higher the selling price will be
```
