```python
import pandas as pd
```

```python
df=pd.read_excel('NZ_Admin_JOBS.xlsx')
```

```python
df.columns = ['Job_Title', 'Job_Link', 'Employer', 'Location','Time Of Post','Classification']
```

```python
pd.set_option('display.max_colwidth',None)
```

```python
df[df['Location'].str.contains(',')]
```

```python
df[['Location','City']]=df['Location'].str.split(':',expand=True)[[1,2]]
```

```python
df['Location']=df['Location'].str.split('area',expand=True)[0]
```

```python
def deduplication(x):
    trim=x.strip()
    index = (trim+trim).find(trim,1)
    if index !=-1:
        return trim[:index]
    else:
        return trim
```

```python
df['Location']=df['Location'].apply(deduplication)
```

```python
df['City'].fillna(df['Location'],inplace=True)
```

```python
df['City']=df['City'].apply(deduplication)
```

```python
df[['posted time','featurned at']] = df['Time Of Post'].str.split(',', expand = True)[[0,2]]
```

```python
def apply_salary(x):
    if 'classification:' in x:
        return 'unknown'
    else:
        return x
```

```python
df['posted time']=df['posted time'].str.extract('(\d+[a-z])')
```

```python
df[df['posted time'].str.contains('m')]
```

```python
df['posted time']=df['posted time'].astype('string')
```

```python
def apply_posted_time(x):
    if 'd' in x:
        return int(x[:-1])
    if 'm' in x:
        return 'Less than one hour'
    if 'h' in x:
        return 'Less than one day'
    else:
        return 0
```

```python
df['posted time']=df['posted time'].apply(apply_posted_time)
```

```python
df['salary']=df[df['salary'].str.contains('\d', na=False)]['salary'].str.replace(' to ','-',regex=True)
```

```python
df['salary'].fillna('unknown')
```

```python
df['salary']=df[df['salary'].str.contains('\d', na=False)]['salary'].str.replace('\d\%','',regex=True)
```

```python
df['salary']=df[df['salary'].str.contains('\d', na=False)]['salary'].str.replace(',','',regex=True)
```

```python
df[['low salary','high salary']] = df['salary'].str.split('-', n=1, expand = True).fillna('unknown')
```

```python
df['low salary'] = df['low salary'].str.extract('(\d+\.\d+|\d+ \d+|\d+k|\d+)')
df['high salary'] = df['high salary'].str.extract('(\d+\.\d+|\d+ \d+|\d+k|\d+)')
```

```python
df['per hr/annum'] = df['salary'].str.split('per', expand = True)[1]
```

```python
df['per hr/annum'].fillna('unknown')
```

```python
df['per hr/annum']=df['per hr/annum'].astype('string')
```

```python
def apply_per_hr(x):
    if x.str.contains('\dd'):
        return 'hour'
    if x.str.contains('\ddddd'):
        return 'annum'
    else:
        return 0
```

```python
df['per hr/annum']=df['per hr/annum'].apply(apply_per_hr)
```
