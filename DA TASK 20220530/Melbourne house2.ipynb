```python
df['Date']=pd.to_datetime(df['Date'],format='%d/%m/%Y',errors='coerce')
```

##### Since a property will at least can park their car offstreet, we will fill in N/A in Car column with 1

```python
df['Car']=df['Car'].fillna('1')
```

##### For easy number reading, property price will be shown in $Million

```python
df['Price']=(df['Price']/1000000).round(2)
df.rename(columns = {'Price':'Price in $M'}, inplace = True)
```

#### since postcode,number of bedroom/bathroom, number of carpark and landsize are all integer number, the data type of these column will change to integer

```python
df[['Postcode','Bedroom2','Bathroom','Car','Landsize']]=df[['Postcode','Bedroom2','Bathroom','Car','Landsize']].astype(int)
```

#### Since Building Area,YearBuilt and CouncilArea,Lattitude,Longtitude and Property count do not have close relationship with property price,so these columns will be dropped

```python
df.drop(['BuildingArea','YearBuilt','CouncilArea','Lattitude','Longtitude','Propertycount'],axis=1,inplace=True)
```

```python
df.head()
```

#### Since the region classification for South-eastern and Southern suburb is not easy to understand, we will then combine these two suburbs before proceed our study.

```python
df['Regionname']= df['Regionname'].replace(['South-Eastern Metropolitan'],['Southern Metropolitan'])
df
```

### 1. quantity sold in each type of properties among different regions

```python
df_region_number = df.groupby(['Regionname','Type']).size().reset_index(name="Count")
df_region_number=df_region_number.sort_values(by='Count', ascending=False)
df_region_number
```

```python
sns.set(rc={"figure.figsize":(10, 5)}) 
sns.barplot(data = df_region_number,x = 'Regionname',y = 'Count',hue='Type')
plt.xticks(rotation = 90)
```

### Southern Metropolitan has the biggest quantities sold among all regions. Study will be focused on Southern metropolitan area to see the possible relationship among price with other factors.
* House type is selected since it has a bigger data set

#### Firstly will be looking at how each suburbs contribute to the total house sold in southern Metropolitan region

```python
df_H=df[(df['Type']=='h')]
df_H
```

```python
df_S=df_H[(df['Regionname']=='Southern Metropolitan')]
df_S
```

```python
df_Suburb= df_S.groupby(['Suburb']).size().reset_index(name="Count")
df_Suburb=df_Suburb.sort_values(by='Count', ascending=False)
df_Suburb.head()
```

```python
sns.set(rc={"figure.figsize":(30, 10)}) 
sns.barplot(data = df_Suburb,x = 'Suburb',y = 'Count')
plt.xticks(rotation = 90)
```

#### Bentleigh East, Balwyn North, Glen Iris and Kew has the biggest number of houses sold. we will be next looking at their price information.

```python
df_BBGK=df_H.loc[df_H['Suburb'].isin(['Bentleigh East','Balwyn North','Glen Iris','Kew'])]
df_BBGK.head()
```

#### Now I will be looking at the price information in these 4 suburbs

```python
df_BBGK.groupby('Suburb').agg({'Price in $M': ['mean','median', 'min', 'max']})
```

#### is there any relationship between landsize and house price in these suburbs?

```python
sns.set(rc={"figure.figsize":(65, 8)})
sns.lmplot(x = "Landsize", y = "Price in $M", data = df_BBGK,hue="Suburb",height=3.5, aspect=4.5)
plt.xlim(200,800) #as from the beginning of this study, we know that 75% of the land size are under 675, here we take 800 as the upper limit for data
plt.ylim(1,5)# as most house prices will be under $5M, we will use $5M as the upper limit
plt.show()
```

#### Price seems to have linear relationship with Landsize in Kew,Glen Iris and Bentleigh East, but not Balwyn North.Let's take closer look on these four suburbs, if number of rooms will affect the house price?

```python
sns.set(rc={"figure.figsize":(35, 8)})
sns.lmplot(x = "Rooms", y = "Price in $M", data = df_BBGK,hue="Suburb",height=3.5, aspect=4.5)
plt.show()
```

#### It's quite clear that it's a linear relationship between number of rooms and house price in Bawlyn North, Glen Iris and especially Kew. It's not very clear for Bentleigh East.

Landsize and number of rooms has a positive relationship with house price in Glen Iris and Kew. Landsize seem to affect more then number of rooms in Bentleigh east, whereas number of rooms affect more than landsize to house price isn Bawlyn North.
